```python
# 8 . Create a new coulmn and try to find our a diffrence between order date and shipment date
```

```python
o.head()
```

```python
o['Order Date'] = pd.to_datetime(o['Order Date'])
o['Ship Date'] = pd.to_datetime(o['Ship Date'])
```

```python
o['Time Difference']=o['Ship Date']-o['Order Date']
```

```python
o.head()
```

```python
# 9 . based on question number 8 find out for which order id we have shipment duration more than 10 days

o[["Order ID","Time Difference"]][o["Time Difference"].dt.days>10].sort_values(by="Time Difference",ascending=False)
```

```python
# 10 . Try to find out a list of a returned order which sihpment duration was more then 15 days and
# find out that region manager as well 

df=o[o["Time Difference"].dt.days>15] # here we filtered rows in orders dataframe which have time difference >15
df1=df.merge(u,on="Region")
df1.iloc[:,23:].sort_values(by="Time Difference",ascending=False)
```

```python
# 11 . Group by region and find out which region is more profitable

reg_profits=o.groupby("Region")["Profit"].sum()
print(f'Profit per region is given as follows:')
reg_profits

# Central region is more profitable followed by East and South then Western region.
```
