```python
# 1. Read this dataset in pandas
import pandas as pd

df=pd.read_csv("FitBit data.csv")
df
```

```python
# 3. convert all the dates avaible in dataset to timestamp format in pandas

df["ActivityDate"]=pd.to_datetime(df["ActivityDate"])
```

```python
df.head()
```

```python
# 4 . Find out in this data that how many unique id's we have 

df['Id'].nunique()
```

```python
# 5 . which id is one of the active id that you have in whole dataset 

a = df[["TotalDistance","Id"]].groupby("Id").sum().reset_index()
b = a.sort_values(by = "TotalDistance",ascending = False,ignore_index = True)
b["Id"][0]
```

```python
# 6 . how many of them have not logged there activity find out in terms of number of ids 

df[df["LoggedActivitiesDistance"]==0]["Id"].nunique()
```

```python
# 7 . Find out who is the laziest person id that we have in dataset 

Dist_tot=df[["TotalDistance","Id"]].groupby("Id").sum().reset_index()
dist_sort=Dist_tot.sort_values(by="TotalDistance",ignore_index=True)
dist_sort["Id"][0]
```

```python
# 8 . Explore over an internet that how much calories burn is required for a healthy person and 
# find out how many healthy person we have in our dataset.

# Most female adults need 1,600–2,200 calories per day, while adult males need 2,200–3,000 calories per day.
# However gender is not provided in dataset we will consider average of both gender calories i.e. 2600
# person is healthy if he burns calories >= 2600
# person is unhealthy if he burns calories < 2600

def calorie_check(a):
    if a>= 2600:
        return "Healthy"
    else:
        return "Unhealthy"
df["health_status"]=df["Calories"].apply(calorie_check)
len(df[df["health_status"]=="Healthy"])
```

```python
# 9. how many person are not a regular person with respect to activity try to find out those 

len(df[df["TotalDistance"]==0])
```

```python
# 10 . who is the third most active person in this dataset find out those in pandas. 

most_active=df[["Id","TotalDistance"]].groupby("Id").sum().reset_index()
most_active_sorted=most_active.sort_values(by="TotalDistance",ascending=False).reset_index()
most_active_sorted["Id"][2]
```

```python
# 11 . who is the 5th most laziest person avilable in dataset find it out 

df.groupby("Id")["TotalSteps"].sum().sort_values().head(5)
```

```python
# 12 . what is a total cumulative calories burn for a person find out 

df[["Id","Calories"]].groupby("Id").sum().sort_values(by="Calories",ascending=False)
```
